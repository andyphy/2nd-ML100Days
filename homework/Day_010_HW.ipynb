```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# 設定 data_path
dir_data = './data/'
```

```python
# 讀取資料檔
f_app_train = os.path.join(dir_data, 'application_train.csv')
app_train = pd.read_csv(f_app_train)
app_train.shape
```

```python
# 將只有兩種值的類別型欄位, 做 Label Encoder, 計算相關係數時讓這些欄位可以被包含在內
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le_count = 0

# 檢查每一個 column
for col in app_train:
    if app_train[col].dtype == 'object':
        # 如果只有兩種值的類別型欄位
        if len(list(app_train[col].unique())) <= 2:
            # 就做 Label Encoder, 以加入相關係數檢查
            app_train[col] = le.fit_transform(app_train[col]) 
            le_count += 1
print(app_train.shape)
print('%d columns were label encoded.' % le_count)
app_train.head()
```

```python
# 受雇日數為異常值的資料, 另外設一個欄位記錄, 並將異常的日數轉成空值 (np.nan)
app_train['DAYS_EMPLOYED_ANOM'] = app_train["DAYS_EMPLOYED"] == 365243
app_train['DAYS_EMPLOYED'].replace({365243: np.nan}, inplace = True)

# 出生日數 (DAYS_BIRTH) 取絕對值 
app_train['DAYS_BIRTH'] = abs(app_train['DAYS_BIRTH'])
```

### 相關係數
一樣，pandas 很貼心地讓我們可以非常容易計算相關係數

```python
# 觀察相關係數
app_train.corr()['TARGET'] 
```

## 練習時間
列出目標 (TARGET) 與所有欄位之間相關係數，數值最大以及最小各 15 個

通過相關係數的結果觀察有興趣的欄位與 TARGET 或其他欄位的相關係數，並嘗試找出有趣的訊息
- 最好的方式當然是畫圖，舉例來說，我們知道  EXT_SOURCE_3 這個欄位和 TARGET 之間的相關係數是 -0.178919 (在已經這個資料集已經是最負的了！)，那我們可以 EXT_SOURCE_3  為 x 軸， TARGET 為 y 軸，把資料給畫出來

```python
#列出目標 (TARGET) 與所有欄位之間相關係數，數值最大以及最小各 15 個
corr_TARGET = app_train.corr()['TARGET'].sort_values()
```

```python
# official answer
print('Most Positive Correlations:\n', corr_TARGET.tail(16), '\n\n')
print('Most Negative Correlations:\n', corr_TARGET.head(15))
```

```python
# my way 
corr_min15 = corr_TARGET[:15]
corr_count = len(corr_TARGET)
corr_max15 = corr_TARGET[corr_count-16:corr_count]
print('Most Positive Correlations:\n', corr_max15, '\n\n')
print('Most Negative Correlations:\n', corr_min15)
```

```python
new_list = list(corr_min15.index)+list(corr_max15.index)
app_train_max_min = app_train[new_list]
for col in app_train_max_min:
    app_train_max_min.boxplot(column=col, by='TARGET')
    plt.suptitle('')
    plt.show()
```

```python
for col in app_train_max_min:
    if col != 'DAYS_EMPLOYED_ANOM':
        app_train_max_min.plot.scatter('TARGET', col)
        plt.suptitle('')
        plt.show()
```
